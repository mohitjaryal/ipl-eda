## Missing values & duplicates

```python
# Missing values summary
missing = ipl.isnull().sum().sort_values(ascending=False)
missing = missing[missing > 0]
missing
```

```python
# duplicates
dups = ipl.duplicated().sum()
dups
```

###  Simple univariate analysis: numeric columns (hist + boxplot)

```python
num_cols = ipl.select_dtypes(include=[np.number]).columns.tolist()
print("Numeric columns:", num_cols)

for col in num_cols:
    series = ipl[col].dropna()
    if len(series)==0:
        continue
    fig, axes = plt.subplots(1,2, figsize=(12,4))
    # histogram
    axes[0].hist(series, bins=30)
    axes[0].set_title(f"Histogram: {col}")
    # boxplot
    sns.boxplot(x=series, ax=axes[1])
    axes[1].set_title(f"Boxplot: {col}")
    plt.tight_layout()
    plt.show()
```

### Simple univariate analysis: categorical columns (value counts + barplot)

```python
cat_cols = ipl.select_dtypes(include=['category','object']).columns.to_list()
cat_cols

for col in cat_cols:
    counts = ipl[col].value_counts(dropna=False)
    display(counts.head(20))
    plt.figure(figsize=(8,3))
    sns.countplot(y=col, data=ipl, order=counts.index[:20])
    # sns.countplot(
    #     y=col, 
    #     data=ipl.dropna(subset=[col]), 
    #     order=counts.index[:20]
    # )
    plt.title(f'Top Category for {col}')
    plt.show()
```

### Correlation heatmap (numeric)

```python
if len(num_cols) >= 2:
    corr = ipl[num_cols].corr()
    plt.figure(figsize=(10,10))
    sns.heatmap(corr,annot=True, cmap='coolwarm',square=True )
    plt.title('Correlation matrix (numeric)')
    plt.show()
```

### Pair Plot

```python
sns.pairplot(ipl[num_cols], diag_kind='kde')
plt.suptitle("Pairwise Relationships (Scatter Matrix)", y=1.02)
plt.show()
```

### Violin Plot

```python
# Example: compare distribution of 'Age' by 'Gender'
sns.violinplot(x='team1', y='team2', data=ipl)
plt.title("Violin Plot: Age vs Gender")
plt.show()
```

### Line Plot

```python
sns.lineplot(x='season', y='win_by_runs', data=ipl)
plt.title("Trend Over Time")
plt.show()
```
